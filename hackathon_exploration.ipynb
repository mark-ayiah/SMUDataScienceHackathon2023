```python
# Importing Libraries
import pandas as pd
import numpy as np
import geopandas
%config InlineBackend.figure_format ='retina'
from fuzzywuzzy import fuzz, process
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading in Data
whr = pd.read_csv("data/whrdata.csv")
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
world = world[(world.pop_est>0) & (world.name!="Antarctica")] # Removing Antarctica
worldvalues = pd.read_csv("data/worldvalues.csv")
```

```python
# Text-mining country names
whr["fuzzynames"] = whr["Country name"].apply(lambda x: " ".join(
    x.lower() for x in x.split())).str.replace("\([^)]*\)", "").str.replace(
    '[^\w\s]','').str.replace("university of", "").str.replace("college", "").str.replace(
    "university", "").str.strip()
world["fuzzynames"] = world["name"].apply(lambda x: " ".join(
    x.lower() for x in x.split())).str.replace("\([^)]*\)", "").str.replace(
    '[^\w\s]','').str.replace("university of", "").str.replace("college", "").str.replace(
    "university", "").str.strip()

# Defining fuzzy match process
def fuzzy_match(name, choices):
    match = process.extractOne(name, choices)
    return match[0]

for index, row in whr.iterrows():
    country_name = row['fuzzynames']
    best_match = fuzzy_match(country_name, world['fuzzynames'].values)
    whr.at[index, 'best_match'] = best_match

# Fixing discrepancies & merging with geo-dataset
whr.at[26, "best_match"] = "taiwan"
whr = whr.drop(labels=[24,36,41,58,81,129], axis = 0)
whr = whr.drop(columns = ["Country name", "fuzzynames"]).rename(columns={"best_match":"name"})
wrld = world.merge(whr, how = 'left', right_on = 'name', left_on = "fuzzynames").drop(columns = ["name_y", "fuzzynames"]).rename(columns={"name_x":"Country Name"})
```

```python
# Plotting ladder score heatmap
ax = wrld.plot(column='Ladder score', legend = True, figsize=(12, 10), 
          cmap = 'plasma', missing_kwds={'color': 'lightgray',
                                          "edgecolor": "white",
                                          "hatch": "///",
                                          "label": "Missing values"}, 
          legend_kwds={"label": "Happiness Ladder Score (2022)", "orientation": "horizontal"});
ax.set_axis_off();
#plt.savefig('happyladderheatmap.jpg', dpi = 600)
```
